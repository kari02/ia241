## 2.1 What are the common job titles?

```python
from collections import Counter

import xlrd
book=xlrd.open_workbook('jobs.xls')
sheet=book.sheet_by_name("Sheet1")

job_title_list=[]
company_list=[]
post_date_list=[]
location_list=[]
min_salary_list=[]
duty_list=[]
skills_list=[]
exp_list=[]
edu_list=[]
URL_list=[]

for i in range(sheet.nrows):
    Job_Title,Company,Posted_Date,Job_Location,Minimal_Salary,Job_Duty,Required_Skills,Year_of_Experience,Minimal_Education,URL=sheet.row_values(i)
    if i !=0:   
        job_title_list.append(Job_Title)
        company_list.append(Company)
        post_date_list.append(Posted_Date)
        location_list.append(Job_Location)
        min_salary_list.append(Minimal_Salary)
        duty_list.append(Job_Duty)
        skills_list.append(Required_Skills)
        exp_list.append(Year_of_Experience)
        edu_list.append(Minimal_Education)
        URL_list.append(URL)
        
count_result=Counter(job_title_list)
    
for word, count in count_result.most_common(5):
    print(word, count)
```

```python
import matplotlib.pyplot as plt

result=Counter(job_title_list)

plt.pie(result.values(),labels=result.keys())
plt.show()
```

The most common job titles, from the data collected, are Business Intelligence Analyst and Intelligence Analyst. Both titles are used four times from different job postings. The rest of the given job titles vary, but are each used only once; therefore, we only included the top five titles. We used a pie chart for this question to represent the many different job titles.

## 2.8 What are the common job duties?

```python
for duty in df['Job_Duty']:
    print (duty)
```

```python
duties =''
for duty in df['Job_Duty']:
    duties = duty + duties
print (duties)
```

```python
import nltk

import xlwt        
from collections import Counter        
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))

book = xlwt.Workbook() # create a new excel file
sheet_test = book.add_sheet('word_count') # add a new sheet
i = 0
sheet_test.write(i,0,'word') # write the header of the first column
sheet_test.write(i,1,'count') # write the header of the second column
sheet_test.write(i,2,'ratio') # write the header of the third column
```

```python
word_list = [i for i in duties.lower().split() if i not in stop]
word_total = word_list.__len__()

count_result =  Counter(word_list)
for result in count_result.most_common(20):
    i = i+1 
    sheet_test.write(i,0,result[0])
    sheet_test.write(i,1,result[1])
    sheet_test.write(i,2,(result[1]/word_total))

book.save('Job_Duties.xls')# define the location of your excel file
```
